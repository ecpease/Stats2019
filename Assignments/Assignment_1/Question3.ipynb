# c

```python
def get_Y(X, eps):
    Y = -1 + 0.5*X + eps
    return Y
```

```python
Y = get_Y(X, eps)
Y
```

```python
print("The length of Y is ", len(Y))
```

B_o and B_1 in the model refer to -1 and 0.5 respectively

# d

```python
fig, ax = plt.subplots()
plt.scatter(X, Y)
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

### The relationship appears as a linear trend. The data points are clustered near (0, -1 )

# e

```python
fig, ax = plt.subplots()
plt.scatter(X, Y)
ax.set_xlabel("X")
ax.set_ylabel("Y")

z = np.polyfit(X, Y, 1)
p = np.poly1d(z)
plt.plot(X, p(X),"r--")
# the line equation:
print("y=%.6fX+(%.6f)"%(z[0],z[1]))
```

### The relationship here remains to appear as a linear trend, with a slope of approximately 0.5 but not equal to 0.5, and an intercept very close to 1, but not equal to 1 exactly.  The data points are more clustered near (0, -1 ).  Both the slope and intercept are very close to the original function's slope and intercept of 0.5 and -1, but they are not exactly correct.  This could be due to random error in the data.

```python
X = np.random.normal(0, 1, 100) # mu, sigma, count
```

```python
eps_05 = np.random.normal(0, .05, 100)
```

```python
Y_05 = get_Y(X, eps_05)
```

```python
fig, ax = plt.subplots()
plt.scatter(X, Y_05)
ax.set_xlabel("X")
ax.set_ylabel("Y")

z = np.polyfit(X, Y_05, 1)
p = np.poly1d(z)
plt.plot(X, p(X),"r--")

print("y=%.6fX+(%.6f)"%(z[0],z[1]))
```

```python
X2 = X**2
X2.shape
```

```python
x3 = []
for i in range(len(X)):
    x3.append([X[i],X2[i]])
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x3, Y_05, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x3, Y_05)
```

```python
reg.score(x3, Y_05)
```

```python
import statsmodels.formula.api as sm
from statsmodels.api import OLS
```

```python
X = np.append(arr = np.ones((len(x3),1)).astype(int), values=x3, axis=1)
```

```python
regressor_OLS = sm.OLS(endog=Y_05, exog=x3).fit()
```

```python
regressor_OLS.summary()
```

In repeating the process with a smaller error (epsilon), the model score increased dramatically, from 0.83 to 0.99.  The R-squared decreased from 0.499 to 0.453, but the adjusted R-squared decreased from 0.489 to 0.442.
