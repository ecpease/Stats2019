```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# a

```python
X = np.random.normal(0, 1, 100) # mu, sigma, count
X
```

```python
X_mu = 0
X_sigma = 1

count, bins, ignored = plt.hist(X, 30, density=True)
plt.plot(bins, 1/(X_sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - X_mu)**2 / (2 * X_sigma**2) ), linewidth=2, color='r')
plt.show()
```

# c

```python
def get_Y(X, eps):
    Y = -1 + 0.5*X + eps
    return Y
```

```python
Y = get_Y(X, eps)
Y
```

```python
print("The length of Y is ", len(Y))
```

B_o and B_1 in the model refer to -1 and 0.5 respectively

# d

```python
fig, ax = plt.subplots()
plt.scatter(X, Y)
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

### The relationship appears to be a linear trend.  The data points are more clustered near (0, -1 )

# e

```python
fig, ax = plt.subplots()
plt.scatter(X, Y)
ax.set_xlabel("X")
ax.set_ylabel("Y")

z = np.polyfit(X, Y, 1)
p = np.poly1d(z)
plt.plot(X, p(X),"r--")
# the line equation:
print("y=%.6fX+(%.6f)"%(z[0],z[1]))
```

### The relationship here remains to appear as a linear trend, with a slope of 0.471033 and an intercept of -1.0278.  The data points are more clustered near (0, -1 ).  Both the slope and intercept are very close to the original function's slope and intercept of 0.5 and -1, but they are not exactly correct.  This could be due to random error in the data.
